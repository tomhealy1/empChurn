## Visualisation

```python
%matplotlib inline
import matplotlib.pyplot as plt
pd.crosstab(hr.function,hr.left).plot(kind='bar')
plt.title('Turnover Frequency for Function')
plt.xlabel('Function')
plt.ylabel('Frequency of Turnover')
plt.savefig('department_bar_chart')
```

The above graph shows us the the frequency of stays and leavers. We can see function can be a good feature to use in any prediction model

```python
table=pd.crosstab(hr.salary, hr.left)
table.div(table.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True)
plt.title('Stacked Bar Chart of Salary Level vs Turnover')
plt.xlabel('Salary Level')
plt.ylabel('Proportion of Employees')
plt.savefig('salary_bar_chart')
```

So shock of shocks!! People on lower salaries were more likely to leave also. Again, we can see that salary may be good feature in a prediction model.

```python
### Assign oun our X (independant variable) and y (dependant variable)
```

```python
hr_vars=hr.columns.values.tolist()
y=['left']
X=[i for i in hr_vars if i not in y]
```

```python
X
```

### Building out Model using Logistic regression and RFE

```python
model = LogisticRegression()
rfe = RFE(model, 5)
rfe = rfe.fit(hr[X], hr[y])
print(rfe.support_)
print(rfe.ranking_)
```

```python
X
```

```python
# make a new var with the RFE columns 
#model_cols=['satisfaction_level', 'last_evaluation','time_spend_company', 'Work_accident', 'promotion_last_5years', 'function_RandD','function_accounting',
# 'function_hr','salary_high',
# 'salary_low']
model_cols=['satisfaction_level','Work_accident','promotion_last_5years', 'function_RandD','salary_high']
```

```python
# X comprises of our indepedent vars and y our output/dependent variable
X = hr[model_cols]
y=hr['left']
```

```python
#Here we are splitting our test sets and training sets
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.3, random_state=0)
```

## Build our Logistic Regression Model

```python
log_mod = LogisticRegression()
log_mod.fit(X_train, y_train)
```

```python
print('Logistic regression accuracy: {:.3f}'.format(accuracy_score(y_test, log_mod.predict(X_test))))
```

## Random Forest 

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
print('Random Forest Accuracy: {:.3f}'.format(accuracy_score(y_test, rf.predict(X_test))))
```

## Cross Validation

```python
from sklearn import model_selection
from sklearn.model_selection import cross_val_score
kfold = model_selection.KFold(n_splits=10, random_state=7)
modelCV = RandomForestClassifier()
scoring = 'accuracy'
results = model_selection.cross_val_score(modelCV, X_train, y_train, cv=kfold, scoring=scoring)
print("10-fold cross validation average accuracy: %.3f" % (results.mean()))
```

The average accuracy is close to the RF moeld so we can say our model generalises well

## Precision and Recall

```python
print(classification_report(y_test, log_mod.predict(X_test)))
```

```python
## Confusion Matrix for Logit model
logreg_y_pred = log_mod.predict(X_test)
logreg_cm = metrics.confusion_matrix(logreg_y_pred, y_test, [1,0])
sns.heatmap(logreg_cm, annot=True, fmt='.2f',xticklabels = ["Left", "Stayed"] , yticklabels = ["Left", "Stayed"] )
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.title('Logistic Regression')
plt.savefig('logistic_regression')
```

```python
print(classification_report(y_test, rf.predict(X_test)))
```

```python
y_pred = rf.predict(X_test)
from sklearn.metrics import confusion_matrix
import seaborn as sns
forest_cm = metrics.confusion_matrix(y_pred, y_test, [1,0])
sns.heatmap(forest_cm, annot=True, fmt='.2f',xticklabels = ["Left", "Stayed"] , yticklabels = ["Left", "Stayed"] )
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.title('Random Forest')
plt.savefig('random_forest')
```

```python
Recall: When someone leaves, how often does our get it right?
Our has correctly us 70% (704/1038) so our recall is around 70%.

Precision: When our model predicts someone will leave, how often does that happen.  87% (704/804) 
```

## ROC Curve

```python
logit_roc_auc = roc_auc_score(y_test, log_mod.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, log_mod.predict_proba(X_test)[:,1])

rf_roc_auc = roc_auc_score(y_test, rf.predict(X_test))
rf_fpr, rf_tpr, rf_thresholds = roc_curve(y_test, rf.predict_proba(X_test)[:,1])

plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot(rf_fpr, rf_tpr, label='Random Forest (area = %0.2f)' % rf_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('ROC')
plt.show()
```

```python
#Feature importance in descending order of importance
feature_labels = np.array(['satisfaction_level', 'Work_accident', 'promotion_last_5years', 
      'department_RandD', 'salary_high'])
importance = rf.feature_importances_
feature_indexes_by_importance = importance.argsort()[::-1]
for index in feature_indexes_by_importance:
    print('{}-{:.2f}%'.format(feature_labels[index], (importance[index] *100.0)))
```
